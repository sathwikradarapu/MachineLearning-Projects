calculating cumulative sum of ballscore grouping by matchid and adding it as new column score

```python
deliveries_df['Score'] = deliveries_df[['match_id','Ball_score']].groupby('match_id').cumsum()['Ball_score']
```

adding new columns to the match_df target-left,balls-remaining,player-dismissed,wickets-fallen,current-runrate,required-runrate,result

```python
# New columns which has target 
deliveries_df['target_left'] = (deliveries_df['total_runs'] + 1) - deliveries_df['Score']
```

```python
deliveries_df
```

```python
# 1 over 6 balls 
# ipl has 20 over match 20 * 6 = 120 
deliveries_df['Remaining Balls'] = (120 - ( (deliveries_df['over'] - 1)*6 + deliveries_df['ball']))
```

```python
deliveries_df
```

```python
deliveries_df['player_dismissed'] = deliveries_df['player_dismissed'].fillna('0') 
```

```python
deliveries_df
```

```python
deliveries_df['player_dismissed'] = deliveries_df['player_dismissed'].apply(lambda x:x if x == '0' else '1').astype('int64')
```

```python
deliveries_df
```

```python
deliveries_df['Wickets'] = deliveries_df[['match_id','player_dismissed']].groupby('match_id').cumsum()['player_dismissed'].values
```

```python
deliveries_df
```

```python
deliveries_df['Wickets'] = 10 - deliveries_df['Wickets'] 
```

```python
deliveries_df
```

```python
deliveries_df['crr'] = (deliveries_df['Score'])*6/(120 - deliveries_df['Remaining Balls'])
```

```python
deliveries_df['rrr'] = (deliveries_df['target_left'])*6/(deliveries_df['Remaining Balls'])
```

```python
# Creating a column result where 1 means won 0 means lost 
def result(row):
    if row['batting_team'] == row['winner']: 
        return 1
    else: 
        return 0
```

```python
deliveries_df['result'] = deliveries_df.apply(result,axis=1)
```
